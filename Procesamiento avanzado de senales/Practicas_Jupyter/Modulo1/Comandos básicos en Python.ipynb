```python
#To read audio files we need to use scipy

from scipy.io import wavfile

#to check the help otion, you have to use the help command and the name function

help(wavfile.read)
```

```python
fs,data=wavfile.read('speech.wav')
```

```python
# Sampling frequency
fs
```

```python
# vector of data
data
```

```python
# size of data
data.size
```

```python
#time 
second=(data.size)/fs
```

```python
second
```

```python
# plot the audio signal
t=np.arange(data.size)/fs
plt.plot(t,data)
plt.xlabel('time')
plt.ylabel('Amplitude')
```

```python
# plot only a range
y=data[4000:6000]
t_y=t[4000:6000]
plt.plot(t_y,y)
plt.xlabel('time')
plt.ylabel('Amplitude')
```

```python
#Play a sound
import IPython
IPython.display.Audio(data, rate=fs)
```

```python
# we can obtain the maximum value of the signal
np.max(abs(y))
```

```python
# save a new file 
wavfile.write('test.wav',fs,y)
```
